4. Add RNA seq expression

```python
import pandas as pd
import database.db as db
import database.models as models
from database.db_manager import add_rna_seq_data
import utils.data_tidier as dt

rna_seq_file = "all_data/gene_name_update/Xe_seedlings_31_07_25_v5C DESeq2 normalised counts table.csv"

rna_seq_df = pd.read_csv(rna_seq_file)


annotationfile = "all_data/gene_name_update/new_genes_gos_etc.csv"
annotation_df = pd.read_csv(annotationfile, sep=";")
new_genes = annotation_df['SeqName'].tolist()
print(new_genes)
print(len(new_genes))


new_rna_seq_df = rna_seq_df[rna_seq_df['Unnamed: 0'].isin(new_genes)]

rna_seq_genes = new_rna_seq_df['Unnamed: 0'].tolist()

# Check if all new genes have RNA seq data
missing_genes = set(new_genes) - set(rna_seq_genes)
print("Genes missing RNA seq data:", missing_genes)


# convert RNA seq data to long format
long_df = dt.transform_to_long(new_rna_seq_df)
long_df.to_csv('all_data/gene_name_update/Xe_seedlings_31_07_25_v5C DESeq2 normalised counts table_tidy', index=False)
print(long_df.head())

df = dt.add_log2(long_df) # add log2 transformation
df = dt.format_time_points(df) # format timepoint
df.rename(columns={"expression": "normalised_expression"}, inplace=True)
print(df.head())
df.to_csv("all_data/gene_name_update/Xe_seedlings_31_07_25_v5C DESeq2 normalised counts table_tidy_for_db.csv", index=False)

species_id = "X. elegans"
experiment_name = "xe_seedlings_time_course"
rna_seq_data = df
add_rna_seq_data(rna_seq_data, species_id, experiment_name )
```
